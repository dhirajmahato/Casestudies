## label Ecoding where there is order or ranking in some categorical data

```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder

# Sample DataFrame with a categorical column
data = {'Education': ['High School', 'Bachelor', 'Master', 'Ph.D', 'Bachelor']}
df = pd.DataFrame(data)

# Create a LabelEncoder object
label_encoder = LabelEncoder()

# Apply label encoding to the 'Education' column
df['Education_encoded'] = label_encoder.fit_transform(df['Education'])

print(df)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans

# Generate synthetic data for clustering
data, ground_truth_labels = make_blobs(n_samples=300, centers=4, random_state=42)

# Create a K-Means clustering model with the desired number of clusters (4 in this case)
kmeans = KMeans(n_clusters=4, random_state=42)

# Fit the K-Means model to the data
kmeans.fit(data)

# Get cluster assignments for each data point
predicted_labels = kmeans.labels_

# Plot the original data points and the cluster centers
plt.figure(figsize=(10, 6))
plt.scatter(data[:, 0], data[:, 1], c=ground_truth_labels, cmap='viridis', marker='o', s=50, label='True Labels')
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], c='red', marker='x', s=200, label='Cluster Centers')
plt.legend()
plt.title('K-Means Clustering')
plt.show()
```
